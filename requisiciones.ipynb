```python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
        (" ", "_"),
        ("ñ","n"),
        ("(", ""),
        (")", "")
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.lower(), b.lower()).lower()
    return s
```
